```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.basketball-reference.com/teams/CHI/1996_games.html"
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
table = soup.find_all(id="games")
```

```python
head=soup.find(class_="thead")
column_names_raw=[head.text for item in head][0]
column_names_polished=column_names_raw.replace("\n",",").split(",")[2:-1]
print(column_names_polished)
```

```python
games=[]
    
for i in range(len(table)):
    
    game=[]
    
    for td in table[i].find_all("td"):
        game.append(td.text)

        games.append(game) 

# df=pd.DataFrame(games, columns=column_names_polished).set_index("Date")
```

```python
n = []
for i in range(len(game)):
    num = i/13
    if num % 1 == 0:
        n.append(i)  
```

```python
from ast import literal_eval
data = []
for i in range(len(n)-1):
    clean = []
    clean.append(game[n[i]:n[i+1]])
    data.append(clean)

games = []
for d in data:
    newest = str(d).lstrip('[').rstrip(']')
    newest = list(literal_eval(newest))
    games.append(newest)
```

```python
df=pd.DataFrame(games, columns=column_names_polished)
```

```python
col = ['Date', 'Drop1', 'Box_Score', '@Home', 'Opponent', 'Outcome',
       'drop', 'Bulls_Points', 'Opp_Points', 'W', 'L', 'Streak', 'Notes']
```

```python
df.columns = col
df
```
