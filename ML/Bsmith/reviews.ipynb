```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Here we are retrieving the adjectives of each review to see the sentiment
for review in df.Review:
  if type(review) == str:
    doc = nlp(review)
    descriptive_term = []
    for token in doc:
      if token.pos_ == 'ADJ':
          descriptive_term.append(token)
    print("Review:", review)
    print("Term:",descriptive_term)
```

```python
# Here we are also picking up intesifiers such as "very", "super", and more. 
for review in df.Review:
  if type(review) == str:
    doc = nlp(review)
    descriptive_term = []
    for token in doc:
      if token.pos_ == 'ADJ':
          prepend = ''
          for child in token.children:
              if child.pos_ != 'ADV':
                continue            
              prepend += child.text + ' '
          descriptive_term.append(prepend + token.text)

    print("Review",review)
    print("Term:",descriptive_term)
```

```python
aspects = []
for review in df.Review:
  if type(review) == str:
    doc = nlp(review)
    descriptive_term = []
    target = []
    for token in doc:
        if token.pos_ == 'NOUN':
            target.append(token.text)
        if token.pos_ == 'ADJ':
          prepend = ''
          for child in token.children:
              if child.pos_ != 'ADV':
                continue            
              prepend += child.text + ' '
          descriptive_term.append(prepend + token.text)

    aspects.append({'Aspect': target,
                    'Description': descriptive_term })
```

```python
for aspect in aspects:
    print(aspect)
```

```python
# Now lets create a model to use Sentiment Analysis

from textblob import TextBlob

polarity = []
subjectivity = []
for aspect in aspects:
    clean = str(aspect['Description']).strip("[']")
    cleaner = clean.replace("'", '')
    cleanest = cleaner.replace(",", '')
    sentiment = TextBlob(cleanest).sentiment
    aspect['Sentiment'] = sentiment
    polarity.append(sentiment.polarity)
    subjectivity.append(sentiment.subjectivity)
```

```python
for i, row in df.iterrows():
    if type(row.Review) != str:
        df = df.drop([i])
df
```

```python
df['Sentiment'] = df['Rating'].map({5:'Positive', 4:'Positive', 3:'Negative', 2:'Negative', 1:'Negative'})
```

```python
# Subjectivity is whether the review was more opinionated or more factual ranging from [0,1]. 1 being more opinionated
# Polarity is measuring how positive a review is ranging from [-1, 1] where -1 is negative and 1 is positive. 0 is neutral
df['Polarity'] = polarity
df['Subjectivity'] = subjectivity
df
```

```python
def getAnalysis(score):
  if score < .12:
    return 'Negative'
  else:
    return 'Positive'
```

```python
df['Prediction'] = df['Polarity'].apply(getAnalysis)
df
```

```python
import nltk
from nltk.corpus import stopwords
from  nltk.stem import SnowballStemmer
```

```python
stop_words = stopwords.words('english')
stemmer = SnowballStemmer('english')
lemma = nltk.wordnet.WordNetLemmatizer()
stop_words.remove('not')
```

```python
# Function for preprocessing tokens the way we want to

def preprocess(text, stem=False, lem=False):
    tokens = []
    for token in text.split():
        token = token.lower()
        token = token.replace(".", '')
        if token not in stop_words:
            if lem:
                tokens.append(lemma.lemmatize(token))
            elif stem:
                tokens.append(stemmer.stem(token))
            else:
                tokens.append(token)
    return " ".join(tokens)
```

```python
# Here we are converting the words into numbers

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils.np_utils import to_categorical

tk = Tokenizer(num_words=20,
               filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
               lower=True,
               split=" ")
tk.fit_on_texts(train_df.Review)

train_seq = tk.texts_to_sequences(train_df.Review)
test_seq = tk.texts_to_sequences(test_df.Review)
```

```python
seq_lengths = df['Review'].apply(lambda x: len(x.split(' ')))
seq_lengths.describe()
```

```python
# We are not truncating any words but padding with zeros instead because these reviews do not contain many words
# This is to prevent the loss of information
maxlength = 25
train_seq_trunc = pad_sequences(train_seq, maxlen=maxlength)
test_seq_trunc = pad_sequences(test_seq, maxlen=maxlength)
```

```python
# We are diving a little bit into layers and ANN

from keras import models, layers

model = models.Sequential()
model.add(layers.Embedding(25, 5, input_length=maxlength))
model.add(layers.Flatten())
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam'
                , loss='binary_crossentropy'
                , metrics=['accuracy'])
model.summary()
```

```python
epochs = 20
history = model.fit(train_seq_trunc, train_df.Positive, epochs=epochs, verbose=0, validation_split=0.5)
```

```python
loss, accuracy = model.evaluate(test_seq_trunc, test_df.Positive)
```

```python
test_df['Predictions'] = model.predict(test_seq_trunc)
```

```python
# Plotting function to compare validation and training set
def eval_metric(history, metric_name):
    metric = history.history[metric_name]
    val_metric = history.history['val_' + metric_name]

    e = range(1, epochs + 1)

    plt.plot(e, metric, 'go', label='Train ' + metric_name)
    plt.plot(e, val_metric, 'b', label='Validation ' + metric_name)
    plt.legend()
    plt.show()
```
