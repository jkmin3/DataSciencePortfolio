```python
# This is just one game
participants = []
for row in match_detail['info']['participants']:
    participants_row = {}
    participants_row['summoner'] = row['summonerName']
    participants_row['champion'] = row['championId']
    participants_row['spell1'] = row['summoner1Id']
    participants_row['spell2'] = row['summoner2Id']
    participants_row['kills'] = row['kills']
    participants_row['deaths'] = row['deaths']
    participants_row['assists'] = row['assists']
    participants_row['totalDamageDealt'] = row['totalDamageDealt']
    participants_row['goldEarned'] = row['goldEarned']
    participants_row['champLevel'] = row['champLevel']
    participants_row['totalMinionsKilled'] = row['totalMinionsKilled']
    participants_row['item0'] = row['item0']
    participants_row['item1'] = row['item1']
    participants_row['item2'] = row['item2']
    participants_row['item3'] = row['item3']
    participants_row['item4'] = row['item4']
    participants_row['item5'] = row['item5']
    participants_row['item6'] = row['item6']
    participants_row['win'] = row['win']
    participants.append(participants_row)
df = pd.DataFrame(participants)
df
```

```python
# check league's latest version
latest = watcher.data_dragon.versions_for_region(na)['n']['champion']
# Lets get some champions static information
static_champ_list = watcher.data_dragon.champions(latest, False, 'en_US')

# champ static list data to dict for looking up
champ_dict = {}
for key in static_champ_list['data']:
    row = static_champ_list['data'][key]
    champ_dict[row['key']] = row['id']

champ_game = []
for row in participants:
    champ_game.append(champ_dict[str(row['champion'])])
```

```python
# This is so that we can get the name of the spells 

static_spell_list = watcher.data_dragon.summoner_spells(latest)

spell_dict = {}
for spell in static_spell_list['data']:
    row = static_spell_list['data'][spell]
    spell_dict[row['key']] = row['name']

spell_one = []
spell_two = []
for row in participants:
    spell_one.append(spell_dict[str(row['spell1'])])
    spell_two.append(spell_dict[str(row['spell2'])])
```

```python
# This is so that our items have names
static_item_list = watcher.data_dragon.items(latest)

item_dict = {}
for num in static_item_list['data']:
    row = static_item_list['data'][num]
    item_dict[num] = row['name']

# Not too sure on how to make this more compact
item_zero = []
item_one = []
item_two = []
item_three = []
item_four = []
item_five = []
item_six = []
for row in participants:
    zero = str(row['item0'])
    one = str(row['item1'])
    two = str(row['item2'])
    three = str(row['item3'])
    four = str(row['item4'])
    five = str(row['item5'])
    six = str(row['item6'])
    if zero == '0':
        item_zero.append('None')
    else:
        item_zero.append(item_dict[zero])
    if one == '0':
        item_one.append('None')
    else:
        item_one.append(item_dict[one])
    if two == '0':
        item_two.append('None')
    else:
        item_two.append(item_dict[two])
    if three == '0':
        item_three.append('None')
    else:
        item_three.append(item_dict[three])
    if four == '0':
        item_four.append('None')
    else:
        item_four.append(item_dict[four])
    if five == '0':
        item_five.append('None')
    else:
        item_five.append(item_dict[five])
    if six == '0':
        item_six.append('None')
    else:
        item_six.append(item_dict[six])
```

```python
# Updating dataframe
df['champion'] = champ_game
df['spell1'] = spell_one
df['spell2'] = spell_two
df['item0'] = item_zero
df['item1'] = item_one
df['item2'] = item_two
df['item3'] = item_three
df['item4'] = item_four
df['item5'] = item_five
df['item6'] = item_six
```

```python
tables = {}
tables['champions'] = (
    '''CREATE TABLE champions(
        champid INT,
        champname VARCHAR(20),
        title VARCHAR(30),
        type VARCHAR(30),
        attackrange INT,
        passive TEXT,
        qspell TEXT,
        wspell TEXT,
        espell TEXT,
        rspell TEXT,
        qtip TEXT,
        wtip TEXT,
        etip TEXT,
        rtip TEXT,
        lore TEXT
    )'''
)
tables['summoners'] = (
    '''CREATE TABLE summoners(
        summonerid INT,
        summonername VARCHAR(35),
        summonerrank VARCHAR(15),
        wins INT,
        losses INT,
        veteran BOOLEAN,
        freshblood BOOLEAN,
        inactive BOOLEAN,
        hotstreak BOOLEAN
    )'''
)
tables['games'] = (
    '''CREATE TABLE games(
        gameid VARCHAR(40),
        summoner VARCHAR(35), 
        gamerank VARCHAR(15),
        champion INT,
        lane VARCHAR(20),
        spell1 INT,
        spell2 INT,
        kills INT,
        deaths INT,
        assists INT,
        totalDamageDealt INT,
        goldEarned INT,
        champLevel INT,
        csPermin FLOAT,
        item0 INT,
        item1 INT,
        item2 INT,
        item3 INT,
        item4 INT,
        item5 INT,
        item6 INT,
        runes VARCHAR(20),
        win BOOLEAN,
        time INT
    )'''
)
tables['items'] = (
    '''CREATE TABLE items(
        itemid INT,
        item VARCHAR(35)
    )'''
)
tables['spells'] = (
    '''CREATE TABLE spells(
        spellid INT,
        spell VARCHAR(20)
    )'''
)

tables['runes'] = (
    '''CREATE TABLE runes(
        runeid INT,
        rune VARCHAR(20)
    )'''
)
```

```python
# This is creating all of my tables and it prints the errors nicely if I already have a table, so that I can keep the tables dictionary the same
connection = mysql.connector.connect(host=endpoint, database=db_name, user=username, password=password)
cursor = connection.cursor()
for table_name in tables:
    table_description = tables[table_name]
    try:
        print("Creating table {}: ".format(table_name), end='')
        cursor.execute(table_description)
    except mysql.connector.Error as err:
        if err.errno == mysql.connector.errorcode.ER_TABLE_EXISTS_ERROR:
            print("already exists.")
        else:
            print(err.msg)
    else:
        print("OK")

cursor.close()
connection.close()
```
